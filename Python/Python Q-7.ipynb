```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import seaborn as sns 
```

```python
#Open the pokedex.json downloaded after uploaded into jupyter notebook 
with open(r'C:\Users\Lenovo\Downloads\q4.json','r',encoding='utf-8') as file:
    # Load its content and make a new dictionary
    data = json.load(file)

    #Order the data by loading into a data frame df_json
df_json = pd.DataFrame(data)
df_json
```

**Query-1 : Get all the Earth meteorites that fell before the year 2000**

```python
df1 = df_json[["name","fall","year"]]
```

```python
filtered_df = df1.loc[(df1['year'] < '2000-01-01T00:00:00.000')
                     & (df1['year'] < '2000-01-01T00:00:00.000')]
# Display
filtered_df[["name","fall","year"]]
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(20, 40))

# Add x-axis and y-axis
ax.plot(filtered_df['name'],
        filtered_df['year'],
        color='purple')

# Set title and labels for axes
ax.set(xlabel="name",
       ylabel="year",
       title="Earth Meteoerites that fell into the year 2000")

plt.show()
```

**Query-2 : Get all the earth meteorites co-ordinates who fell before the year 1970**
