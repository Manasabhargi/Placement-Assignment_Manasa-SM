```python
import json
import pandas as pd
```

```python
# Open the pokedex.json downloaded after uploaded into jupyter notebook 
with open("pokedex.json") as file:
    # Load its content and make a new dictionary
    data = json.load(file)

    #Order the data by loading into a data frame df_json
df_json = pd.DataFrame(data)
df_json
```

```python
#Then the structed and ordered data is dumped into a csv file and exported using to_excel
df_json.to_excel(r'‪Exported Pokemon Data.xlsx', index=False)
```
